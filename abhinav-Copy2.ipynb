```python
import numpy as np
import matplotlib.pyplot as plt

city1 = np.array([100,120,85,90,110,95])
city2 = np.array([80,75,60,95,85,90])
city3 = np.array([150,140,135,160,155,170])

# city1 = [100,120,85,90,110,95]
# city2 = [80,75,60,95,85,90]
# city3 = [150,140,135,160,155,170]

def totalRainfall(city1, city2, city3):
    total1=0
    total2=0
    total3=0

    for i in city1:
        total1 = total1 + i

    for j in city2:
        total2 = total2 + j

    for k in city3:
        total3 = total3 + k

    print(f"\ntotal rainfall in city x: {total1}")
    print(f"total rainfall in city y: {total2}")
    print(f"total rainfall in city z: {total3}\n")

    print(f"average monthly rainfall in city x: {total1/6}")
    print(f"average monthly rainfall in city y: {total2/6}")
    print(f"average monthly rainfall in city z: {total3/6}\n")

    print(f"total rainfall on all citites : {total1+total2+total3}")
    print(f"average rainfall across all cities : {(total1+total2+total3)/3}")
    print(f"average monthly rainfall : {(total1+total2+total3)/6}")

totalRainfall(city1, city2, city3)

```

```python
plt.plot(city1)
plt.title("rainfall trend of city x")
plt.xlabel("months")
plt.ylabel("rainfall in mm")
plt.legend("x")
```

```python
plt.plot(city2)
plt.title("rainfall trend of city y")
plt.xlabel("months")
plt.ylabel("rainfall in mm")
plt.legend("y")
```

```python
plt.plot(city3)
plt.title("rainfall trend of city z")
plt.xlabel("months")
plt.ylabel("rainfall in mm")
plt.legend("z")
```
